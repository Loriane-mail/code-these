```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'histologie_mice.xlsx'
df = pd.read_excel(file_path)

# Function to convert markers to numerical values
def marker_to_value(marker):
    values = {'+++': 5, '++': 5, '+': 2, '+/-': 2, '-': 1}
    return values.get(marker.strip(), 0)  # Strip whitespace

# Define a color mapping for each marker
color_mapping = {'+++': 'darkred', '++': 'darkred', '+': 'lightcoral', '+/-': 'lightcoral', '-': 'mistyrose'}

# Function to plot pie chart
def plot_pie_chart(data, title):
    labels = ['heart', 'liver', 'brain', 'kidney', 'Lung']  # Ensure consistent order
    sizes = [marker_to_value(data[label]) for label in labels]
    colors = [color_mapping[data[label].strip()] for label in labels]
    
    wedges, texts = plt.pie(sizes, labels=labels, colors=colors, startangle=90, counterclock=False)
    plt.setp(wedges, edgecolor='black')  # Add edge color
    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Determine the number of mice
num_mice = len(df)
cols = 3  # Number of columns for subplots
rows = (num_mice + cols - 1) // cols  # Calculate rows needed

# Plot pie charts for each mouse
plt.figure(figsize=(cols * 5, rows * 5))

for i, row in df.iterrows():
    plt.subplot(rows, cols, i + 1)
    data = row[['heart', 'liver', 'brain', 'kidney', 'Lung']]
    plot_pie_chart(data, f"Mouse {row['ID_souris']} ({row['Groupe']})")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'histologie_mice.xlsx'
df = pd.read_excel(file_path)

# Function to convert markers to numerical values for proportion
def marker_to_value(marker):
    values = {'+++': 1, '++': 1, '+': 1, '+/-': 1, '-': 1}  # All markers have the same value for proportion
    return values.get(marker.strip(), 0)  # Strip whitespace

# Define a color mapping for each marker
color_mapping = {'+++': 'darkred', '++': 'darkred', '+': 'lightcoral', '+/-': 'lightcoral', '-': 'mistyrose'}

# Function to plot pie chart
def plot_pie_chart(data, title):
    labels = ['heart', 'liver', 'brain', 'kidney', 'Lung']  # Ensure consistent order
    sizes = [marker_to_value(data[label]) for label in labels]
    colors = [color_mapping[data[label].strip()] for label in labels]
    
    wedges, texts = plt.pie(sizes, colors=colors,labels=labels,labeldistance=0.5, startangle=90, counterclock=False)
    plt.setp(wedges, edgecolor='black')  # Add edge color
    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Determine the number of mice
num_mice = len(df)
cols = 3  # Number of columns for subplots
rows = (num_mice + cols - 1) // cols  # Calculate rows needed

# Plot and save pie charts for each mouse
plt.figure(figsize=(cols * 5, rows * 5))

for i, row in df.iterrows():
    plt.subplot(rows, cols, i + 1)
    data = row[['heart', 'liver', 'brain', 'kidney', 'Lung']]
    title = f"Mouse {row['ID_souris']} ({row['Groupe']})"
    plot_pie_chart(data, title)
    # Save individual plot
    plt.figure()
    plot_pie_chart(data, title)
    filename = f"Mouse_{row['ID_souris']}_Group_{row['Groupe']}.png"
    plt.savefig(filename)
    plt.close()

# Save the overall figure with all subplots
plt.tight_layout()
plt.savefig('Overall_Figure.png')
plt.show()
```
