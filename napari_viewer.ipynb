```python
import napari
from tifffile import imread
import pandas as pd
import numpy as np
import numpy as np
from glob import glob
import scipy.io
import matplotlib.pyplot as plt
from natsort import natsorted
import os
import pandas as pd
import napari

def mat_to_csv(matfile):
    filename = matfile.split(os.extsep)[0]+".csv"
    mat = scipy.io.loadmat(matfile)["tab_param"]
    print(mat)
    for_df = []

    track_id = 0
    iterator = 0
    for i in range(mat.shape[1]):
        intensities = mat[3:mat.shape[0]:8,i]
        x = mat[1:mat.shape[0]:8,i]
        y = mat[2:mat.shape[0]:8,i]
        times = np.where(intensities != 0.0)[0]

        x = x[times]
        y = y[times]

        for k in range(len(x)):
            for_df.append([iterator,track_id,x[k],y[k],times[k]])
            iterator+=1

        track_id+=1


    df = pd.DataFrame(for_df, columns = ["ID","TRACK_ID","X","Y","T"])
    return df

stack = imread('stack.tif')
print(f'Stack of shape {stack.shape} successfully loaded.')
mat1 = 'Puit1 Jurkat RL-HT 561 +biot 4D6 R-HT 646 cell6_w1Quad-TIRF561_t1.TIF_tab_param.mat'
mat2 = 'Puit1 Jurkat RL-HT 561 +biot 4D6 R-HT 646 cell6_w2Quad-TIRF642_t1.TIF_tab_param.mat' #None 

df = mat_to_csv(mat1)
data = df.loc[
    :, ['TRACK_ID', 'T', 'X', 'Y']
].to_numpy()

if mat2 is not None:
    df2 =  mat_to_csv(mat2)
    data2 = df2.loc[
        :, ['TRACK_ID', 'T', 'X', 'Y']
    ].to_numpy()

viewer = napari.Viewer()
viewer.add_image(stack,name='Fluo',channel_axis=1)
viewer.add_tracks(data, name='642')
if mat2 is not None:
    viewer.add_tracks(data2, name='488')
napari.run()
```
